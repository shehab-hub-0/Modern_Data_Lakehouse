```python
import pandas as pd
from faker import Faker
import random
import uuid

# تهيئة مكتبة Faker
fake = Faker()

# تحديد عدد السطور اللي محتاجها
num_rows = 1000 

# القوائم الخاصة ببياناتك بناءً على الصور
event_types = ['page_view', 'checkout_start', 'purchase', 'add_to_cart', 'remove_from_cart', 'product_view']
page_urls = ['explore', 'app/tag/blog', 'explore/list/app', 'search/categories/tag', 'blog/explore/search', 'tags/blog']
categories = ['cat_electronics', 'cat_clothing', 'cat_beauty', 'cat_sports', 'cat_home']
referrers = ['facebook', 'google', 'instagram', 'email', 'twitter']
devices = ['desktop', 'mobile', 'tablet']

data = []

for i in range(num_rows):
    # إنشاء بيانات مشابهة للهيكل اللي في الصور
    event_id = f"evt_{i+1:06d}"
    customer_id = f"cust_{random.randint(10000, 99999)}"
    session_id = f"sess_{uuid.uuid4().hex[:8]}"
    
    # اختيار عشوائي للبيانات
    event_type = random.choice(event_types)
    product_id = f"prod_{random.randint(1000, 9999)}" if event_type in ['purchase', 'add_to_cart', 'product_view'] else None
    category_id = random.choice(categories) if product_id else None
    
    row = {
        "event_id": event_id,
        "customer_id": customer_id,
        "session_id": session_id,
        "event_type": event_type,
        "event_timestamp": fake.date_time_between(start_date='-30d', end_date='now').strftime('%Y-%m-%d %H:%M:%S'),
        "page_url": random.choice(page_urls),
        "product_id": product_id,
        "category_id": category_id,
        "referrer_source": random.choice(referrers),
        "device_type": random.choice(devices),
        "user_agent": fake.user_agent(),
        "ip_address": fake.ipv4()
    }
    data.append(row)

# تحويل البيانات لـ DataFrame
df = pd.DataFrame(data)

# حفظ الملف بصيغة CSV
file_name = "raw_customer_events.csv"
df.to_csv(file_name, index=False)

print(f"تم إنشاء الملف بنجاح: {file_name}")
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# عدد السطور المطلوب
num_rows = 1000

data = []

# قوائم المعرفات (Ids) عشان الداتا تطلع متسقة
warehouses = [f"wh_{i:03d}" for i in range(1, 6)]
suppliers = [f"supp_{i:03d}" for i in range(1, 6)]

for i in range(num_rows):
    snapshot_id = f"snap_{i+1:06d}"
    product_id = f"prod_{random.randint(1000, 9999)}"
    warehouse_id = random.choice(warehouses)
    supplier_id = random.choice(suppliers)
    
    # الكميات (المنطق الحسابي)
    on_hand = random.randint(30, 500)
    reserved = random.randint(5, 40)
    available = on_hand - reserved # الحساب التلقائي زي الصورة
    
    reorder_point = random.randint(20, 100)
    reorder_quantity = random.randint(50, 200)
    
    unit_cost = round(random.uniform(50.0, 1000.0), 2)
    
    row = {
        "snapshot_id": snapshot_id,
        "product_id": product_id,
        "warehouse_id": warehouse_id,
        "snapshot_date": fake.date_between(start_date='-60d', end_date='today').strftime('%Y-%m-%d'),
        "quantity_on_hand": on_hand,
        "quantity_reserved": reserved,
        "quantity_available": available,
        "reorder_point": reorder_point,
        "reorder_quantity": reorder_quantity,
        "supplier_id": supplier_id,
        "last_received_date": fake.date_between(start_date='-90d', end_date='-10d').strftime('%Y-%m-%d'),
        "unit_cost": unit_cost
    }
    data.append(row)

# تحويل لـ DataFrame وحفظ الملف
df_inventory = pd.DataFrame(data)
df_inventory.to_csv("inventory_snapshots.csv", index=False)

print("تم إنشاء ملف المخازن بنجاح!")
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# عدد السطور المطلوب
num_rows = 1000

data = []

# القوائم بناءً على الصور
payment_methods = ['google_pay', 'apple_pay', 'bank_transfer', 'paypal', 'debit_card']
payment_statuses = ['pending', 'failed', 'completed', 'refunded']
currencies = ['USD']
response_codes = ['00', '51', '05', '14', '61', '85'] # أكواد الرد الشائعة في البنوك
billing_countries = ['MG', 'GN', 'ST', 'CR', 'BB', 'UG', 'DZ', 'RU', 'SV', 'ER', 'AR', 'LT', 'FI']

for i in range(num_rows):
    transaction_id = f"txn_{i+1:06d}"
    order_id = f"ord_{random.randint(100000, 999999)}"
    customer_id = f"cust_{random.randint(10000, 99999)}"
    
    amount = round(random.uniform(10.0, 2000.0), 2)
    # حساب رسوم بوابة الدفع كنسبة بسيطة (زي ما في الصورة حوالي 0.2% - 0.5%)
    gateway_fee = round(amount * random.uniform(0.002, 0.005), 2)
    
    status = random.choice(payment_statuses)
    
    row = {
        "transaction_id": transaction_id,
        "order_id": order_id,
        "customer_id": customer_id,
        "payment_method": random.choice(payment_methods),
        "payment_status": status,
        "amount": amount,
        "currency": "USD",
        "transaction_timestamp": fake.date_time_between(start_date='-30d', end_date='now').strftime('%Y-%m-%d %H:%M:%S'),
        "processor_response_code": random.choice(response_codes),
        "gateway_fee": gateway_fee,
        "merchant_id": f"merch_{random.randint(1000, 9999)}",
        "billing_country": random.choice(billing_countries),
        "risk_score": random.randint(10, 99)
    }
    data.append(row)

# تحويل لـ DataFrame وحفظ الملف
df_payments = pd.DataFrame(data)
df_payments.to_csv("raw_payment_transactions.csv", index=False)

print("تم إنشاء ملف المعاملات المالية بنجاح!")
```

```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

# عدد السطور المطلوب
num_rows = 1000

data = []

# القوائم بناءً على بيانات الصور
carriers = ['FedEx', 'UPS', 'DHL', 'Aramex', 'USPS']
shipping_methods = ['standard', 'express', 'overnight', 'economy']
shipping_statuses = ['shipped', 'in_transit', 'delivered', 'pending_pickup', 'returned']

for i in range(num_rows):
    shipment_id = f"ship_{i+1:06d}"
    order_id = f"ord_{random.randint(100000, 999999)}"
    
    # تواريخ منطقية: تاريخ الشحن وتاريخ الوصول المتوقع
    ship_date = fake.date_time_between(start_date='-20d', end_date='now')
    est_delivery = ship_date + pd.Timedelta(days=random.randint(2, 7))
    
    row = {
        "shipment_id": shipment_id,
        "order_id": order_id,
        "carrier_name": random.choice(carriers),
        "tracking_number": f"TRK{random.randint(100000000, 999999999)}",
        "shipping_method": random.choice(shipping_methods),
        "shipping_status": random.choice(shipping_statuses),
        "shipping_date": ship_date.strftime('%Y-%m-%d %H:%M:%S'),
        "estimated_delivery_date": est_delivery.strftime('%Y-%m-%d %H:%M:%S'),
        "actual_delivery_date": (est_delivery if random.random() > 0.3 else None), # 70% تم التوصيل
        "shipping_address_id": f"addr_{random.randint(1000, 9999)}",
        "shipping_cost": round(random.uniform(5.0, 50.0), 2),
        "weight_kg": round(random.uniform(0.5, 20.0), 2)
    }
    data.append(row)

# حفظ الملف
df_shipping = pd.DataFrame(data)
df_shipping.to_csv("row_support_tickets.csv", index=False)

print("تم إنشاء ملف بيانات الشحن بنجاح!")
```
